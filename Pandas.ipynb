```python

import pandas as pd
```

```python
# Series: 1D data
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

```python

# DataFrame: 2D table
data = {
    'Name': ['Ali', 'Sara', 'Zeeshan'],
    'Age': [25, 22, 28],
    'City': ['Lahore', 'Karachi', 'Islamabad']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Access a column
print(df['Name'])

```

```python
# Access a column
print(df['City'])

```

```python

# Access rows by index
print(df.loc[0])    # first row

```

```python
print(df.iloc[1])   # second row
```

```python
# Statistics
print(df['Age'].mean())   # average age
print(df['Age'].sum())    # sum of ages
```

```python
updated_df=df[df['Age'] > 23]
print(updated_df)   # rows where Age > 23
```

```python
# Add a new column
df['Country'] = 'Pakistan'
print(df)
```

```python

# Write DataFrame to CSV
df.to_csv('output.csv', index=False)
```

```python
# Read CSV file
df = pd.read_csv('output.csv')
```
